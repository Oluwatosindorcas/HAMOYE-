```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('FoodBalanceSheets_E_Africa_NOFLAG.csv.xls', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df.describe()
```

```python
#11. Total Animal Fat Produced in 2014:
df_item = df.groupby('Item')

print ('The total sum of animal fat produced in 2014 is', df_item.get_group('Animal fats')['Y2014'].sum())
```

```python
#Total Animal Fat Produced in 2017:

print ('The total sum of animal fat produced in 2017 is', df_item.get_group('Animal fats')['Y2017'].sum())
```

```python
#12. The mean across the dataset for the year 2015
print ('The mean across the dataset for the year 2015 is', df['Y2015'].mean())

#The standard deviation across the dataset for the year 2015
print ('The standard deviation across the dataset for the year 2015 is', df['Y2015'].std())
```

```python
df.head()
```

```python
#13. Total number of data values in year 2016 
total = df['Y2016'].count()

#Sum of missing values in year 2016
null = df['Y2016'].isnull().sum()

#Percentage of missing values in 2016
perc = (null/total)*100
print('sum of missing values in 2016 is', null)
print('percentage of missing values in 2016 is', perc)
```

```python
#14. Correlation of all numeric values
df.corr()

#Correlation between 2014 and Element Code
Corr1 = df['Y2014'].corr(df['Element Code'])
print ("The Correlation between Year 2014 and Element Code is", Corr1)

#Correlation between 2015 and Element Code
Corr2 = df['Y2015'].corr(df['Element Code'])
print ("The Correlation between Year 2015 and Element Code is", Corr2)

#Correlation between 2016 and Element Code
Corr3 = df['Y2016'].corr(df['Element Code'])
print ("The Correlation between Year 2016 and Element Code is", Corr3)

#Correlation between 2017 and Element Code
Corr4 = df['Y2017'].corr(df['Element Code'])
print ("The Correlation between Year 2017 and Element Code is", Corr4)
print('\n')

print("Therefore year 2014 has the highest correlation with Element Code")
```

```python
df_Element = df.groupby('Element')
df_Element.first()
```

```python
#15. The sum of import quantity for each year
print('The sum of import quantity for 2014 is', df_Element.get_group('Import Quantity')['Y2014'].sum())
print('The sum of import quantity for 2015 is', df_Element.get_group('Import Quantity')['Y2015'].sum())
print('The sum of import quantity for 2016 is', df_Element.get_group('Import Quantity')['Y2016'].sum())
print('The sum of import quantity for 2017 is', df_Element.get_group('Import Quantity')['Y2017'].sum())
print('The sum of import quantity for 2018 is', df_Element.get_group('Import Quantity')['Y2018'].sum())
print('\n')
print('The year with the highest sum of import quantity is 2017')
```

```python
#16. Total number of sum of production in 2014
print("The total number of the sum of production in 2014 is",df_Element.get_group('Production')['Y2014'].sum())
```

```python
#17
select_df = df[['Y2018', 'Element']]
print(select_df.groupby('Element').get_group('Fat supply quantity (g/capita/day)').sum())
print(select_df.groupby('Element').get_group('Seed').sum())
print(select_df.groupby('Element').get_group('Protein supply quantity (g/capita/day)').sum())
print(select_df.groupby('Element').get_group('Food supply (kcal/capita/day)').sum())
print(select_df.groupby('Element').get_group('Food supply quantity (kg/capita/yr)').sum())
print('\n')
print('Seed has the third lowest sum in 2018')
```

```python
#19. Total import quantity in algeria in 2018
print("The total import quantity in Algeria in 2018 is", df[df['Area'] == 'Algeria'].groupby('Element').get_group('Import Quantity')['Y2018'].sum())
```
